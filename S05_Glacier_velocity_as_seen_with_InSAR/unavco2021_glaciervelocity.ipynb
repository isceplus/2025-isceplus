#### A brief overview of glaciers
* Glaciers are rivers of ice. They transport mass gained through snowfall to lower elevations where it's lost through melting from above by the atmosphere and below by the oceans or lakes.

* Glaciology (the study of glaciers) can be broken down along many different lines. Examples include the study of surface mass balance (snowfall, wind, and melt) and glacier dynamics. 
* InSAR is most commonly used to study glacier dynamics, though I would argue we have yet to realize the full potential of InSAR for studying glaciers. 
* We're going to focus today on studies of glacier dynamics.

#### Key differences between glacier flow and solid-Earth deformation
Some examples of deformation of the solid-Earth are discussed in this course, so it's useful to ground our discussion of glacier dynamics by relating the fundamentals of each and noting some important distinctions. 

* From a mechanics perspective,
    * Solid-Earth is a solid. Solids support shear stresses.
    * Glacier ice is (often) a fluid. Fluids deform continuously under applied shear stresses.
* Note that we use ‘fluid’ and ‘solid’ to describe mechanical properties not states of matter. So, liquids are fluids but fluids are not necessarily liquids. Fluids can be in any state of matter.
* Ice is a poly-crystalline solid (state of matter) that flows as a viscous fluid (over timescales observable with InSAR).
* Deformation and flow of ice is facilitated by the motion of voids and dislocations (you can think of these as microscopic imperfections) within and between individual ice crystals.



* The dependence of ice flow on creep mechanisms means that we can use InSAR observations from satellites to infer the motion of H2O molecules! That's about *15 orders of magnitude*.
* To make such inferences, we have to understand the connections between what we can observe and what we want to know. 


#### Key descriptors and their relationship to InSAR measurements
* We can often describe things in terms of:
    * motion, which is a “kinematic” description
    * forces that drive motion, which is a “dynamical” description
* We can observe kinematic quantities with InSAR (e.g., surface velocity of glaciers)
* Often want to relate kinematic observations to dynamics (e.g., what creep mechanism is dominant?) 
* Solids vs fluids: Usually interested in different, but related, kinematic (InSAR) observations because the dynamical descriptions differ
    * Solids: 
        * Displacement and deformation (gradient of displacement, related to strain)
        * $\mathrm{strain} \propto \mathrm{stress}$ via Hooke's Law
    * Fluids: 
        * Flow rate (velocity) and rate of deformation (gradient of velocity, related to “strain rate”)
        * $\mathrm{strain\ rate} \propto \mathrm{stress}^n$
        * Thus, viscosity $\eta \propto \mathrm{stress}^{1-n}$
        * $n=1$ is a Newtonian fluid, meaning viscosity is independent of stress (liquid water is an example of a Newtonian fluid)
        * But for glaciers, $2\le n\le 5$, meaning that the viscosity of ice decreases as stress increases; this stress dependence of viscosity is one of things that makes glacier flow so interesting
        * What determines the value of $n$? Creep mechanisms discussed above. 
            - Diffussion creep: $n=1$
            - dislocation creep $n>3$ (probably around $n=4$)
            - "plastic deformation" or failure of ice can result in $n\to\infty$
            - other creep mechanisms can have a range of values
* Value of the stress exponent $n$ creates challenges and opportunities for InSAR observations; let's see how...

#### Influence of ice rheology on InSAR observations

* Rheology of ice varies according to the mechanism that allows ice to creep, the temperature of the ice, and other factors
* All of these can vary in space and are generally sensitive to the rate of deformation (gradient of the velocity field)
* There are a variety of positive feedbacks that can ice to soften in areas of high deformation, which drives further concentration of deformation

* Softer ice and high value of $n$ cause deformation to localize. For example, the steep gradients in the velocity profile in this image are indicative of locally warm ice that has a strong crystallographic preferred orientation (fabric), both of which soften the ice
* Localization of deformation and steep gradients in the velocity profile are simply different ways to saying that the strain rates (defined as the symmetric components of the velocity gradients) are high

* This localization of deformation (i.e., high strain rates) manifests in an interferogram as a high fringe rate 
* Unwrapping is difficult in areas where the fringe rate is high
* In some cases, the deformation rate can be too high to resolve using InSAR

#### Influence of vertical velocity on InSAR observations
* Recall that InSAR measures displacement along the radar line-of-sight (LOS)
* LOS is virtually always oblique to the surface of glaciers
* But glacier velocity is 3D and displacement measured by InSAR has contributions from factors other than glacier flow
* When studying glaciers with InSAR, need to consider the projection of the motion of the glacier surface into the radar LOS.
* Let us consider the mechanisms that can drive vertical 'motion' in InSAR observations
    * Vertical component of glacier velocity -- to a good approximation, ice is incompressible, which means that whenever the horizontal flow of the glacier changes in space, the vertical velocity changes, too. More explicitly, mass conservation requires that $\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z} = 0$ in an incompressible fluid. In an idealized glacier, we expect that $u_z$ is downward in the upstream areas (called 'submergence velocity') and upward in the downstream areas (called 'emergence velocity'). 
    * Vertical velocity should not be confused with changes in the surface elevation of the glacier. When a glacier is in steady-state (mass in = mass out), surface elevation will remain constant but vertical velocity will be non-zero. Most glaciers are not in steady-state and their surface elevation is constantly changing. 
    * Vertical motion due to ocean tides. In the above image from Goldstein et al., 1993, you'll see a line labeled the 'grounding line', the boundary between the grounded ice upstream and the floating ice shelf downstream. The floating ice shelf rises and falls with the ocean tides. This vertical motion manifests in interferogram as shown in the Goldstein image. Since the ocean tides have semi-diurnal and diurnal periods, this motion can have a marked influence on interferograms collected over short timescales.  
         * Vertical motion due to the ocean tides is regularly used to map the grounding line; this is often done by differencing two interferograms collected at different times. Since horizontal flow is *approximately* constant over tidal timescales, this subtracting two interferograms removes most of the signal due to horizontal flow and leaves the difference in vertical displacement. The grounding line is proximal to the upstream extent of the non-zero vertical displacement.
         * Vertical motion observed in InSAR data can also be used to estimate the amplitudes and phases (i.e., relative timing of peak tide) of the ocean tides (see Minchew et al., 2017 for more)
         * Vertical displacement due to ocean tides are nuissance signals when focused on changes in ice thickness or velocity. The effect of ocean tides can be largely removed using a tidal model (e.g., CATS2008 in Antarctica) 
         * Below is an example from the same area as the Goldstein et al., 1993, image above (data from Minchew et al., 2017; note that these data were all collected in the same year). The areas where the observed LOS displacement changes significantly is the floating ice shelf. The grounded ice shows relatively little change in displacement over the observational timeframe. Compare the shape of the floating region to the grounding line shown in Goldstein et al., 1993 (above).
        
<img src="img/rutford_tidal_csk1.png" alt="time series of tidal motion of Rutford Ice Stream" width=800 height=800 />

* Finally, changes in the electromagnetic (EM) properties (permittivity and conductivity) of the glacier near-surface can cause apparent vertical motion in InSAR by changing the effective penetration depth. 
    * Recall from your prepatory courses that penetration depth is defined as $\delta \approx \sqrt{\varepsilon'}/(k_0\varepsilon'')$, where $\varepsilon'$ is the permittivity, $\varepsilon''$ is a loss factor that scales with conductivity, and $k_0$ is the radar wavenumber ($2\pi$/wavelength) in free space. (The dielectric constant is defined as $\varepsilon = \varepsilon' - i\varepsilon''$, where $i^2 = -1$.) 
    * A simple, semi-emperical model for changes in EM properties of snow due to changes in liquid water content $0 \le \nu_w \le 0.2$ is:
        * $\varepsilon' \approx 1.9 + 2\nu_w + \psi\nu_w^{1.3}$
        * $\varepsilon'' = \theta\psi\nu_w^{1.3}$
        * where $\psi = 30/(1 + \theta^2)$, $\theta = f_0/f_r$, $f_0$ is the frequency (in GHz) of the radar in free space, and $f_r = 9$ GHz is a reference frequency. 
    * As an exercise, plot the above functions for radar frequencies from 1-10 GHz to build some intuition as to how does the penetration depth change with liquid water content and radar frequency. 
    * An example of changes in apparent vertical motion due to changes in liquid water content at the glacier near-surface (see Minchew et al., 2015, for a more detailed description):

#### Common noise sources when observing glaciers with InSAR
* Decorrelation of InSAR signals is common, especially during winter and the beginning of melt seasons. 
* Decorrelation can be caused by a variety of factors, but three to be cognizant of are: 
    * Changes in the EM properties, as discussed above
    * Reduced SNR due to low backscatter return in dry snow
    * Volumetric scattering in dry snow
* Let's briefly consider each:

* Changes in EM properties:
    * Most commonly due to melting at the surface of glaciers. Surface melt is an increasingly common occurance on glaciers and ice sheets. Many glaciers outside of the coldest areas of Antarctica experience surface melt from time to time, and many of these areas experience surface melt every summer. 
    * When water is ice, the molecules are held in a regular orientation through H bonding. When water is a liquid, the molecules move around more freely and are better able to rotate in response to an applied EM field (radar). 
    * Thus, the dielectric constant is strongly dependent on liquid water content, causing the radar image to decorrelate during times with the presence of liquid water at the surface changes between radar acquisitions. This typically occurs early in the melt season when melting and refreezing can occur each day at the glacier near surface.
    * Decorrelation due to melting motivates short repeat-pass data acquisitions. As an example, let us consider changes to the correlation of the radar signal over an Icelandic ice cap during the melt season. (In the figure, the left column is data collected during winter when there is no surface melt and the righ column is data collected during early summer when surface melt is becoming more common.)

* Low SNR in dry snow:
    * Dry snow contains a lot of air and is a poor reflector of EM waves. Therefore, most of the transmitted energy from the radar will be lost through scattering and attenuation in the snow pack. Very little will return to the radar. As a result, the SNR will often be low in areas of undisturbed dry snow and thermal noise from the radar will decrease the correlation of InSAR images.
    * In areas of rapid deformation, such as the shear margins in the Goldstein image above, this effect will be mitigated by damage (e.g., crevassing) of the surface. In slower-deforming regions, SNR will generally be low.
    * You will see some examples of this in the lecture on glacier offset fields. 
* Volumetric scattering in dry snow:
    * Backscatter recorded by the radar is a function of scattering from the surface of the glacier and from within the snow that covers the glacier (volume scattering)
    * Very little power is scattered from the surface of dry snow because the EM properties of dry snow are not much different from those of the atmosphere. Volume scatterers generally have lower correlation due to the random path (thus scattering properties) that the EM radiation takes through the volume. 
    * Wetter glacier surfaces will scatter more power from the surface, which can be more stable so long as the liquid water content remains approximately constant between radar acquisitions. 

#### Summary
* Glaciers are among the more exciting and challenging features that can be studied with InSAR
* In essence, glaciers are geological processes playing out over human timescales. 
* Thus, what you learn about InSAR in this UNAVCO course can be applied to glaciers.
* But there are a number of challenges and other factors to keep in mind when studying glaciers. 
* The ones we went over in this lecture include:
    * Relationship between observed surface observation and the properties at depth.
    * Influence of ice rheology on InSAR observations, namely the challenges of unwrapping across rapidly deforming areas like shear margins
    * Multiple mechanisms that can manifest as vertical motion in InSAR observations, from flow to ocean tidal uplift to changes in the EM properties of the glacier surface 
    * Noise sources, which can differ somewhat from those noise sources commonly encountered in solid-Earth applications.
